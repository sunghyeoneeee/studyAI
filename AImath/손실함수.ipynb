```python
import sympy as sp
import numpy as np
a = sp.Symbol("a")
a = sp.Symbol("a")
listx = np.array([3, 6, 9, 3, 6, 9])
listy = np.array([5, 8, 11, 3, 4, 5])

def lof( X, Y, a):
  listx = np.array(X)
  listy = np.array(Y)
  expecta = a*listx
  dista = listy - expecta
  return(sp.expand(sum(dista**2)/len(listx))) # 손실함수
lof1 = lof([3, 6, 9, 3, 6, 9], [5, 8, 11, 3, 4, 5], a)
```

```python
print(lof1.subs(a, 4/3))
print(lof1.subs(a, 2/3))
print(lof1.subs(a, 1))
print(lof1.subs(a, 5/3))
print(lof1.subs(a, 5/6))
print(lof1.subs(a, 1/3))
```

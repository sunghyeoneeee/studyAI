```python
import matplotlib.pyplot as plt

x_data = ['MON', 'TUE', 'WED', 'THR', 'FRI', 'SAT', 'SUN']
y_data = [242, 256, 237, 223, 263, 81, 46]

plt.title('Data', fontsize = 16)
plt.xlabel('Days', fontsize = 12)
plt.ylabel('Human', fontsize = 12)

plt.scatter(x_data[:5], y_data[:5], color = 'hotpink')
plt.plot(x_data, y_data, color = 'pink')
plt.show()
```
